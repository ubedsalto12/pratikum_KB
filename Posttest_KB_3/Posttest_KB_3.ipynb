Meng Import library pandas ,Matplotlib dan seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

membaca file csv menggunakan library pandas

```python
data = pd.read_csv('fitness_class_2212.csv')
data.head(10)
```

membuat pie chart menggunakan library matplotlib dengan membuat variable baru bernama dataCategory menggunakan library pandas
dataCategory: Digunakan sebagai nilai untuk membuat pie chart. Ini adalah hasil dari value_counts() yang mencerminkan berapa kali setiap kategori muncul dalam kolom 'category'.
explode=explode: Menggunakan nilai explode yang telah ditentukan sebelumnya untuk meledakkan beberapa irisan.
labels=dataCategory.index: Menetapkan label untuk setiap irisan pie chart. Dalam hal ini, label diambil dari indeks hasil value_counts(), yaitu nilai unik dalam kolom 'category'.
autopct='%1.1f%%': Menetapkan format untuk menampilkan persentase pada setiap irisan dengan satu desimal.
startangle=100: Memutar pie chart sehingga dimulai pada sudut 100 derajat searah jarum jam dari sumbu x.

```python
explode = (0, 0, 0, 0,0.1,0.1)  # explode the 1st slice
dataCategory = data.category.value_counts()
plt.figure(figsize=(10, 10))
plt.pie(dataCategory,explode=explode, labels=dataCategory.index, autopct='%1.1f%%', startangle=100)
plt.title('Category Pie')
plt.show()
```

newColumnNames = ['Booking ID', 'Month As Member', 'Weight', 'Attended']: Membuat daftar baru yang berisi nama kolom yang ingin Anda tampilkan pada heatmap.

dataCorr = data.corr(): Menghitung matriks korelasi dari dataset data menggunakan fungsi .corr() dari Pandas.

sns.heatmap(dataCorr, cmap='coolwarm', annot=True, center=0, xticklabels=newColumnNames, yticklabels=newColumnNames): Membuat heatmap dari matriks korelasi dataCorr dengan menggunakan Seaborn (sns). Pengaturan warna heatmap menggunakan skema 'coolwarm', nilai-nilai korelasi ditampilkan di dalam sel-sel heatmap (parameter annot=True), dan label sumbu x dan y diatur sesuai dengan newColumnNames.

plt.title('Heatmap'): Menetapkan judul untuk heatmap.

plt.show(): Menampilkan heatmap di jendela grafik.

```python
newColumnNames = ['Booking ID', 'Month As Member', 'Weight', 'Attended']
dataCorr = data.corr()
sns.heatmap(dataCorr, cmap='coolwarm', annot=True, center=0,xticklabels=newColumnNames, yticklabels=newColumnNames)
plt.title('Heatmap Graph ')
plt.show()
```

plt.hist(data['day_of_week'], bins=30, color='red'):

plt.hist(): Fungsi ini digunakan untuk membuat histogram. Data yang digunakan diambil dari kolom 'day_of_week'.
bins=30: Mengatur jumlah bin atau interval pada histogram menjadi 30.
color='red': Mengatur warna histogram menjadi merah.
plt.title('Histogram Hari'):

plt.title(): Digunakan untuk memberikan judul pada plot. Dalam hal ini, judulnya adalah 'Histogram Hari'.
plt.xticks(rotation=90):

plt.xticks(): Mengatur label pada sumbu x. Dalam hal ini, label 'Hari' diputar 90 derajat untuk menghindari tumpang tindih.
plt.ylabel('Frekuensi Kemunculan'):

plt.ylabel(): Memberikan label pada sumbu y, yaitu 'Frekuensi Kemunculan'.
plt.xlabel('Hari'):

plt.xlabel(): Memberikan label pada sumbu x, yaitu 'Hari'.
plt.show():

plt.show(): Menampilkan plot histogram.

```python
plt.hist(data['day_of_week'],bins=30,color='red')
plt.title('Histogram Banyaknya Hari')
plt.xticks(rotation=90)
plt.ylabel('Frekuensi Kemunculan')
plt.xlabel('Hari')
plt.show()
```

```python
x = dataCategory.index  # Mengambil indeks kategori dari dataCategory
y = dataCategory  # Mengambil frekuensi kemunculan kategori dari dataCategory

# Membuat diagram batang dengan kategori (x) dan frekuensinya (y)
plt.bar(x, y, color=['red', 'green', 'blue', 'orange', 'yellow', 'pink'])

# Menyertakan label pada sumbu x dan y serta judul diagram
plt.xlabel("Category")  # Label sumbu x
plt.ylabel("Banyaknya Yang Mengikuti")  # Label sumbu y
plt.title('Diagram Batang Banyaknya Mengikuti Kelas Olahraga Berdasarkan Kategori')  # Judul diagram

plt.show()  # Menampilkan diagram batang
```
