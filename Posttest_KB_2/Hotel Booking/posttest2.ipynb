Menggunakan Liblary pandas,membaca file csv menggunakan pandas dan membuat dua variabel pada variabel kolomInt berfungsi memilih kolom yang bertipe int

```python
import pandas as pd
data = pd.read_csv("hotel_bookings.csv")
kolomAngka = data.select_dtypes(include=['int','float'])
kolomObject = data.select_dtypes(include=['object'])
```

Tampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka. Masukkan informasi tersebut dalam dictionary dengan key sesuai nama informasinya. Tampilkan isi dari dictionary tersebut sebagus mungkin

```python


# Mendapatkan informasi yang diinginkan
jumlah_record = len(kolomAngka)  # Jumlah record (baris)
jumlah_attribute = len(kolomAngka.columns)  # Jumlah attribute (kolom)
nama_attribute = kolomAngka.columns.tolist()  # Nama attribute

# Menyusun informasi dalam dictionary
informasi_dict = {
    'Jumlah Record': jumlah_record,
    'Jumlah Attribute': jumlah_attribute,
    'Label Attribute Bertipe Angka': nama_attribute,
}

# Mencetak isi dictionary dengan format yang lebih baik
print("====================")
print("Informasi DataFrame:")
print("====================")
for key, value in informasi_dict.items():
    if isinstance(value, list):
        print("===================================")
        print(f"{key} ({len(value)}):")
        
        for item in value:
            print(f"  - {item}")
    else:
        print(f"{key}: {value}")
print("===================================")
```
